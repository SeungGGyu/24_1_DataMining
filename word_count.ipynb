```python
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import konlpy
```

```python
df = pd.read_csv(r"C:\Users\qls05\OneDrive\바탕 화면\sos.csv", encoding = 'euc-kr')

disease = df[df['target']=='전염병']
typhoon = df[df['target']=='태풍']
rain = df[df['target']=='호우']
cold = df[df['target']=='한파']
traffic = df[df['target']=='교통']
flood = df[df['target']=='홍수']
snow = df[df['target']=='대설']
hot = df[df['target']=='폭염']
landslide = df[df['target']=='산사태']
etc = df[df['target']=='기타']
```

```python
# disease["송출내용"] = disease["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
typhoon["송출내용"] = typhoon["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
rain["송출내용"] = rain["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
cold["송출내용"] = cold["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
traffic["송출내용"] = traffic["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
flood["송출내용"] = flood["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
snow["송출내용"] = snow["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
hot["송출내용"] = hot["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
landslide["송출내용"] = landslide["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
etc["송출내용"] = etc["송출내용"].str.replace('[^가-힣]', ' ', regex = True)
```

```python
kkma = konlpy.tag.Kkma()
```

```python
# nouns = disease['송출내용'].apply(kkma.nouns)
nouns_typhoon = typhoon['송출내용'].apply(kkma.nouns)
nouns_rain = rain['송출내용'].apply(kkma.nouns)
nouns_cold = cold['송출내용'].apply(kkma.nouns)
nouns_traffic = traffic['송출내용'].apply(kkma.nouns)
nouns_flood = flood['송출내용'].apply(kkma.nouns)
nouns_snow = snow['송출내용'].apply(kkma.nouns)
nouns_hot = hot['송출내용'].apply(kkma.nouns)
nouns_landslide = landslide['송출내용'].apply(kkma.nouns)
nouns_etc = etc['송출내용'].apply(kkma.nouns)
```

```python
# nouns_etc = nouns_etc.explode()
nouns_typhoon = nouns_typhoon.explode()
nouns_rain = nouns_rain.explode()
nouns_cold = nouns_cold.explode()
nouns_traffic = nouns_traffic.explode()
nouns_flood = nouns_flood.explode()
nouns_snow = nouns_snow.explode()
nouns_hot = nouns_hot.explode()
nouns_landslide = nouns_landslide.explode()
nouns_etc = nouns_etc.explode()
```

```python
# df_disease = pd.DataFrame({'word' : nouns_etc})
# df_disease['count'] = df_disease['word'].str.len()
# df_disease = df_disease.query('count >= 2')

df_typhoon = pd.DataFrame({'word' : nouns_typhoon})
df_typhoon['count'] = df_typhoon['word'].str.len()
df_typhoon = df_typhoon.query('count >= 2')

df_rain = pd.DataFrame({'word' : nouns_rain})
df_rain['count'] = df_rain['word'].str.len()
df_rain = df_rain.query('count >= 2')

df_cold = pd.DataFrame({'word' : nouns_cold})
df_cold['count'] = df_cold['word'].str.len()
df_cold = df_cold.query('count >= 2')

df_traffic = pd.DataFrame({'word' : nouns_traffic})
df_traffic['count'] = df_traffic['word'].str.len()
df_traffic = df_traffic.query('count >= 2')

df_flood = pd.DataFrame({'word' : nouns_flood})
df_flood['count'] = df_flood['word'].str.len()
df_flood = df_flood.query('count >= 2')

df_snow = pd.DataFrame({'word' : nouns_snow})
df_snow['count'] = df_snow['word'].str.len()
df_snow = df_snow.query('count >= 2')

df_hot = pd.DataFrame({'word' : nouns_hot})
df_hot['count'] = df_hot['word'].str.len()
df_hot = df_hot.query('count >= 2')

df_landslide = pd.DataFrame({'word' : nouns_landslide})
df_landslide['count'] = df_landslide['word'].str.len()
df_landslide = df_landslide.query('count >= 2')

df_etc = pd.DataFrame({'word' : nouns_etc})
df_etc['count'] = df_etc['word'].str.len()
df_etc = df_etc.query('count >= 2')
```

## count 결과

```python
# df_disease = df_disease.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_typhoon = df_typhoon.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_rain = df_rain.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_cold = df_cold.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_traffic = df_traffic.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_flood = df_flood.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_snow = df_snow.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_hot = df_hot.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_landslide = df_landslide.groupby('word', as_index=False).count().sort_values('count', ascending=False)
df_etc = df_etc.groupby('word', as_index=False).count().sort_values('count', ascending=False)
```
